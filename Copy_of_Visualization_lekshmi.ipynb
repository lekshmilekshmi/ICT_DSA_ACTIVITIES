```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/Sample - Superstore.xls")

# Quick look at the data
print(df.head())
print(df.info())
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="Sales", y="Profit", data=df)
plt.title("Sales vs Profit")
plt.show()
```

```python
# Sales by Category
plt.figure(figsize=(5,4))
sns.barplot(x="Category", y="Sales", data=df, estimator=sum)
plt.title("Total Sales by Category")
plt.show()
```

```python
# Profit by Region
plt.figure(figsize=(5,4))
sns.barplot(x="Region", y="Profit", data=df, estimator=sum)
plt.title("Profit by Region")
plt.show()
```

```python
# Top 10 Sub-Categories by Sales
plt.figure(figsize=(5,4))
subcat_sales = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)
sns.barplot(x=subcat_sales.values, y=subcat_sales.index)
plt.title("Top 10 Sub-Categories by Sales")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(x="Category", y="Sales", data=df)
plt.title("Sales Distribution by Category")
plt.show()
```

```python
# Histogram for Sales
plt.figure(figsize=(5,4))
sns.histplot(df['Sales'], bins=50, kde=True)
plt.title("Distribution of Sales")
plt.show()
```

```python
# Histogram for Profit
plt.figure(figsize=(5,4))
sns.histplot(df['Profit'], bins=50, kde=True, color="orange")
plt.title("Distribution of Profit")
plt.show()
```

```python
# Sales by Region & Category
plt.figure(figsize=(5,4))
sns.barplot(x="Region", y="Sales", hue="Category", data=df, estimator=sum)
plt.title("Total Sales by Region and Category")
plt.show()
```

```python
# Countplot for Category
plt.figure(figsize=(6,4))
sns.countplot(x="Category", data=df, palette="Set2")
plt.title("Count of Orders by Category")
plt.show()
```

```python
# Sales vs Profit by Category
plt.figure(figsize=(5,4))
sns.scatterplot(x="Sales", y="Profit", hue="Category", data=df, alpha=0.7)
plt.title("Sales vs Profit by Category")
plt.show()
```

```python
# Pairplot for numeric variables
sns.pairplot(df[["Sales","Profit","Quantity","Discount"]], diag_kind="kde")
plt.suptitle("Multivariate Relationships", y=1.02)
plt.show()
```
