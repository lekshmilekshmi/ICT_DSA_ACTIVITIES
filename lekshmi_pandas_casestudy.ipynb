```python
#1
import pandas as pd
import numpy as np
url="https://raw.githubusercontent.com/laxmimerit/All-CSV-ML-Data-Files-Download/master/IMDB-Movie-Data.csv"
df2=pd.read_csv(url)
```

```python
df2.duplicated().sum()
df2=df2.drop_duplicates()
df2
```

```python
df2= df2.rename(columns={"Title":"Movie_Title"})
df2
```

```python
unique_genres=set()
for g in df2["Genre"].dropna():
  unique_genres.update(g.split(" , "))
print("\nUnique Genres:",unique_genres)
```

```python
df2["Num_Genres"] = df2["Genre"].apply(lambda x: len(x.split(", ")) if pd.notnull(x) else 0)
df2["Num_Genres"]
```

```python
df2["Revenue (Millions)"] = df2["Revenue (Millions)"].fillna(df2["Revenue (Millions)"].median())
df2["Metascore"] = df2["Metascore"].fillna(df2["Metascore"].median())
df2[["Revenue (Millions)","Metascore"]].isnull().sum()
```

```python
top5_movies = df2[["Movie_Title", "Rating"]].sort_values(by="Rating", ascending=False).head(5)
print(top5_movies)
```

```python
highest_rev = df2.loc[df2["Revenue (Millions)"].idxmax()]
lowest_rev = df2.loc[df2["Revenue (Millions)"].idxmin()]
print("\nHighest Revenue Movie:\n", highest_rev[["Movie_Title", "Revenue (Millions)"]])
print("\nLowest Revenue Movie:\n", lowest_rev[["Movie_Title", "Revenue (Millions)"]])
```

```python
avg_rating_director = df2.groupby("Director")["Rating"].mean().sort_values(ascending=False)
print("\nAverage Rating by Director:\n", avg_rating_director.head())
```

```python
movies_per_year = df2["Year"].value_counts().sort_index()
print("\nMovies per Year:\n", movies_per_year)
```

```python
high_rating_revenue = df2[(df2["Rating"] > 8) & (df2["Revenue (Millions)"] > 100)]
print("\nMovies with Rating > 8 & Revenue > 100M:\n", high_rating_revenue[["Movie_Title", "Rating", "Revenue (Millions)"]])
```

```python
director_group = df2.groupby("Director").agg({"Rating":"mean", "Revenue (Millions)":"mean"}).sort_values(by="Rating", ascending=False)
print("\nDirector-wise Avg Rating & Revenue:\n", director_group.head())
```

```python
year_group = df2.groupby("Year").agg({"Movie_Title":"count", "Rating":"mean"}).rename(columns={"Movie_Title":"Movie_Count"})
print("\nYear-wise Count & Avg Rating:\n", year_group)
```

```python
def categorize_rating(r):
    if r < 6:
        return "Poor"
    elif 6 <= r < 7.5:
        return "Average"
    else:
        return "Excellent"

df2["Rating_Category"] = df2["Rating"].apply(categorize_rating)
print("\nRating Categories:\n", df2["Rating_Category"].value_counts())
```

```python
def movie_score(row):
    return row["Rating"] * 10 + row["Revenue (Millions)"]

df2["High_Score"] = df2.apply(movie_score, axis=1)
print("\nHigh Score Column Added:\n", df2[["Movie_Title", "High_Score"]].head())
```

```python

df2['Profitability'] = df2['Revenue (Millions)'] / df2['Runtime (Minutes)']
top10_profitable = df2.nlargest(10, 'Profitability')[['Movie_Title','Profitability']]
print("\nTop 10 Most Profitable Movies:\n", top10_profitable)
```
