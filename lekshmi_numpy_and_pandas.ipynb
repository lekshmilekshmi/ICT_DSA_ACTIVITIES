```python
# 1.Create a NumPy array

import numpy as np
a = np.arange(10, 50)
print("array:\n",a)
matrix = a.reshape(4, 10)
print("reshaped matrix:\n",matrix)
last_col = matrix[:, -1]
print("last column:\n",last_col)
```

```python
# 2.Create two 3×3 matrices:

import numpy as np
A=np.full((3,3),5)
print("5s matrix:\n",A)
B=np.arange(1,10).reshape(3,3)
print("1 to 9 matrix:\n",B)
C=A+B
print("sum of matrices:\n",C)
D=A*B
print("matrix multiplication:\n",D)
transpose=D.T
print("Transpose:\n",transpose)
```

```python
# 3.Create a NumPy array of 15 random integers between 10 and 100.

import numpy as np
array = np.random.randint(10, 101, 15)
print("array between 10 & 100:",array)
replaced = array > 50
array[replaced] = -1
print("replaced array:",array)
count_replaced = np.sum(replaced)
print("replaced count:\n",count_replaced)

```

```python
# 4.Generate a 5×4 array with random values between 0 and 1 (float).

import numpy as np
arr4 = np.random.rand(5, 4)
print("Array:\n", arr4)

row_sums = arr4.sum(axis=1)
max_row_index = np.argmax(row_sums)
print("\nRow sums:", row_sums)
print("Row with maximum sum (index):", max_row_index)
print("Row with maximum sum:", arr4[max_row_index])


row_min = arr4.min(axis=1, keepdims=True)
row_max = arr4.max(axis=1, keepdims=True)
normalized_arr = (arr4 - row_min) / (row_max - row_min)

print("\nNormalized Array:\n", normalized_arr)
```

```python
# 5.unique elements and duplicates

import numpy as np
arr = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
unique_elements = np.unique(arr)
unique, counts = np.unique(arr, return_counts=True)
print("Unique elements:", unique_elements)
duplicates = unique[counts > 1]
print("Duplicates:", duplicates)
```

```python
# 6. Create a 6×6 NumPy array with values from 1 to 36

import numpy as np
ar=np.arange(1,37).reshape(6,6)
print("6*6 array:\n",ar)
submatrix = ar[1:5, 1:5]
print("middle 4×4 submatrix:\n",submatrix)
even_no = submatrix[submatrix % 2 == 0]
print("even numbers:",even_no)
```

```python
# 7.Generate an array of shape (3, 4) with values from 0 to 11.

import numpy as np
ar1=np.arange(0,12).reshape(3,4)
print("array of 3*4:\n",ar1)
reshape = ar1.reshape(2, 6)
print("reshaped:\n",reshape)
flatt = ar1.flatten()
print("Flattened 1D Array:\n", flatt)
ar2 = flatt.reshape(4, 3)
print("Reshaped back to (4x3) Array:\n", ar2)
```

```python
# 8.Create an array:

import numpy as np
a = np.array([[1, 2],[3, 4]])
a = np.vstack([a, [5, 6]])
a = np.column_stack([a, [7, 8, 9]])
print(a)
```

```python
# 9.

import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])
arr2 = np.concatenate([a, b, c])
print("Joined array:", arr2)

matrix1 = arr2.reshape(3, 3)
print("3x3 matrix:\n", matrix1)

diag_elements = np.diag(matrix1)
print("Diagonal elements:", diag_elements)
```

```python
# 10.Create a NumPy array of integers from 1 to 20.

import numpy as np
arr = np.arange(1, 21)
print("array:", arr)
div = arr[(arr % 3 == 0) | (arr % 5 == 0)]
print("Divisible by 3 or 5:", div)
count = div.size
print("Count:", count)
```

# **pandas**

```python
# 1.

import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic.head(10)
```

```python
#2.
import pandas as pd
missing_values = titanic.isnull().sum()
missing_values
```

```python
#3.
import pandas as pd
survival_by_gender = titanic.groupby("sex")["survived"].mean() * 100
survival_by_gender
```

```python
#4.
import pandas as pd
avg_age = titanic["age"].mean()
print("Average Age:", avg_age)
median_age = titanic["age"].median()
print("Median Age:", median_age)
std_age = titanic["age"].std()
print("Standard Deviation of Age:", std_age)
```

```python
#5.
class_counts = titanic["pclass"].value_counts()
class_counts
```

```python
#6.
import pandas as pd
survived = titanic[(titanic["sex"] == "female") & (titanic["age"] < 18) &(titanic["survived"] == 1)]
survived
```

```python
#7.
titanic['age'].fillna(titanic['age'].median(), inplace=True)
print(titanic['age'].isnull().sum())
```

```python
#9.
top_fares = titanic.sort_values(by="fare", ascending=False).head(5)
top_fares
```

```python
#10.
avg_fare = titanic.groupby(["pclass", "sex"])["fare"].mean()
print(avg_fare)
```

```python
#11.
titanic.rename(columns={'sex': 'gender', 'fare': 'ticket_fare'}, inplace=True)
print(titanic.head())
```

```python
#13.
avg_age_grouped = titanic.groupby(["pclass", "gender"])["age"].mean()
print(avg_age_grouped)
```
